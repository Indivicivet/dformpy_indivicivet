```python
import formpy as fp
import numpy as np
import matplotlib.pyplot as plt
```

# <center> Plotting Options </center>

```python
p=np.linspace(-4,4,23)
x,y=np.meshgrid(p,p)
u=y*np.sin(y)
v=x*np.cos(x)
fp1=fp.form_1(x,y,u,v)
plt.figure(1,figsize=(6,6))
ax=plt.axes()
fp1.plot(ax)
```

```python
w=u*v
fp2=fp.form_2(x,y,w)
plt.figure(2,figsize=(6,6))
ax2=plt.axes()
fp2.plot(ax2)
```

```python
vf=fp.vector_field(x,y,u,v)
plt.figure(3,figsize=(6,6))
ax3=plt.axes()
vf.plot(ax3)
```

```python
p=np.linspace(-4,4,100)
xs,ys=np.meshgrid(p,p)
us=ys*np.sin(ys)
vs=xs*np.cos(xs)
ws=us*vs
fp0=fp.form_0(xs,ys,ws)
plt.figure(4,figsize=(6,6))
ax4=plt.axes()
fp0.plot(ax4)
```

# <center> Zooming Options </center>

```python
p=np.linspace(-4,4,23)
x,y=np.meshgrid(p,p)
u=y*np.sin(y)
v=x*np.cos(x)
fp1=fp.form_1(x,y,u,v)
fp1.give_eqn('y*sin(y)','x*cos(x)')
plt.figure(1,figsize=(6,6))
ax=plt.axes()
fp1.plot(ax)
z_ax, z_fp1 = fp1.zoom((0, 0), zoom=20, dpd=9, inset=True, axis=ax)
```

```python
fp2=fp.form_2(x,y,w)
fp2.give_eqn('x*y*cos(x)*sin(y)')
plt.figure(2,figsize=(6,6))
ax2=plt.axes()
fp2.plot(ax2)
zax2,zfp2=fp2.zoom((0,0),zoom=4,dpd=9,inset=True,axis=ax2)
```

```python
plt.figure(3,figsize=(6,6))
ax3=plt.axes()
fp0.give_eqn('x*y*sin(y)*cos(x)')
fp0.plot(ax3)
zax3,zfp0=fp0.zoom((0,0),zoom=4,dpd=9,inset=True,axis=ax3)
```

```python
plt.figure(4,figsize=(6,6))
ax4=plt.axes()
vf=fp.vector_field(x,y,u,v)
vf.give_eqn('y*sin(y)','x*cos(x)')
vf.plot(ax4)
zax4,zvf=vf.zoom((2,3),zoom=4,dpd=9,inset=True,axis=ax4)
```

# <center> Vector Field Calculus </center>

```python
plt.figure(4,figsize=(6,6))
ax4=plt.axes()
vf=fp.vector_field(x,y,u,v)
vf.give_eqn('y*sin(y)','x*cos(x)')
vf.plot(ax4)
zax4,zvf=vf.DF((2,3),zoom=4,dpd=9,inset=True,axis=ax4)
```

```python
plt.figure(4,figsize=(6,6))
ax4=plt.axes()
vf=fp.vector_field(x,y,u,v)
vf.give_eqn('y*sin(y)','x*cos(x)')
vf.plot(ax4)
zax4,zvf=vf.Div((2,3),zoom=4,dpd=9,inset=True,axis=ax4)
```

```python
plt.figure(4,figsize=(6,6))
ax4=plt.axes()
vf=fp.vector_field(x,y,u,v)
vf.give_eqn('y*sin(y)','x*cos(x)')
vf.plot(ax4)
zax4,zvf=vf.Curl((2,3),zoom=4,dpd=9,inset=True,axis=ax4)
```

# <center> Exterior Algebra </center>

```python
fig5=plt.figure(5,figsize=(18,6))
ax51=fig5.add_subplot(131)
ax52=fig5.add_subplot(132)
ax53=fig5.add_subplot(133)
fp0=fp.form_0(x,y,w)
fp0.plot(ax51)
num_dfp0=fp0.num_ext_d()
num_dfp0.plot(ax52)
fp0.give_eqn('x*y*sin(y)*cos(x)')
ana_dfp0=fp0.ext_d()
ana_dfp0.plot(ax53)
```

```python
fig6=plt.figure(6,figsize=(18,6))
ax61=fig6.add_subplot(131)
ax62=fig6.add_subplot(132)
ax63=fig6.add_subplot(133)
fp1=fp.form_1(x,y,u,v)
fp1.plot(ax61)
num_dfp1=fp1.num_ext_d()
num_dfp1.plot(ax62)
fp1.give_eqn('y*sin(y)','x*cos(x)')
ana_dfp1=fp1.ext_d()
ana_dfp1.plot(ax63)
```

```python
q=np.linspace(-4,4,21)
[x,y]=np.meshgrid(q,q)
u=-y
v=x
fp1=fp.form_1(x,y,u,v)
fig7=plt.figure(7,figsize=(18,6))
ax71=fig7.add_subplot(131)
ax72=fig7.add_subplot(132)
ax73=fig7.add_subplot(133)
fp1.plot(ax71)
num_dfp1=fp1.num_ext_d()
num_dfp1.plot(ax72)
fp1.give_eqn('-y','x')
ana_dfp1=fp1.ext_d()
ana_dfp1.plot(ax73)
```

# <center> Conservative vector fiels </center>

```python
q=np.linspace(-4,4,21)
[x,y]=np.meshgrid(q,q)
u=y
v=x
fp1=fp.form_1(x,y,u,v)
fig8=plt.figure(8,figsize=(18,6))
ax81=fig8.add_subplot(131)
ax82=fig8.add_subplot(132)
ax83=fig8.add_subplot(133)
fp1.plot(ax81)
num_dfp1=fp1.num_ext_d()
num_dfp1.plot(ax82)
fp1.give_eqn('y','x')
ana_dfp1=fp1.ext_d()
ana_dfp1.plot(ax83)
```

```python
q=np.linspace(-4,4,21)
[x,y]=np.meshgrid(q,q)
u=x/((x**2+y**2)**(1.5))
u=y/((x**2+y**2)**(1.5))
fp1=fp.form_1(x,y,u,v)
fig9=plt.figure(9,figsize=(18,6))
ax91=fig9.add_subplot(131)
ax92=fig9.add_subplot(132)
ax93=fig9.add_subplot(133)
fp1.plot(ax91)
num_dfp1=fp1.num_ext_d()
num_dfp1.plot(ax92)
fp1.give_eqn('y','x')
ana_dfp1=fp1.ext_d()
ana_dfp1.plot(ax93)
```
