```python
import formpy as fp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-4,4,100)
r=(np.cosh(x))**(2/3)
plt.plot(x,r)
```

```python
p=np.linspace(-2,2,23)
x,y=np.meshgrid(p,p)
u=x*y
v=np.sin(x*y)
f1=fp.form_1(x,y,u,v,'x*y','sin(x*y)')
f1p = fp.form_1(x, y, u*v, v, 'x*y*sin(x*y)', 'sin(x*y)')
f0=fp.form_0(x, y, u, 'x*y')
f2 = fp.form_2(x, y, v, 'sin(x*y)')
```

```python
f11w=f1.wedge(f1p)
```

```python
ax=plt.axes()
f11w.plot(ax)
```

```python
f10w=f1.wedge(f0)
```
